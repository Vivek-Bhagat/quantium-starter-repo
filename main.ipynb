```python
import pandas as pd
import glob

# Read all CSV files
csv_files = glob.glob('data/daily_sales_data_*.csv')

# Initialize an empty list to store dataframes
dfs = []

# Read each CSV file
for file in csv_files:
    df = pd.read_csv(file)
    # Calculate Sales = price * quantity
    # Remove $ sign from price and convert to float
    df['price'] = df['price'].str.replace('$', '').astype(float)
    df['sales'] = df['price'] * df['quantity']
    # Select only required columns
    df_output = df[['sales', 'date', 'region']].copy()
    # Rename columns to match required format
    df_output.columns = ['Sales', 'Date', 'Region']
    dfs.append(df_output)

# Combine all dataframes
combined_df = pd.concat(dfs, ignore_index=True)

# Save to output file
output_file = 'combined_sales_output.csv'
combined_df.to_csv(output_file, index=False)

print(f'Successfully combined {len(csv_files)} files into {output_file}')
print(f'Total rows: {len(combined_df)}')
print(f'\nFirst few rows:')
print(combined_df.head(10))
print(f'\nData summary:')
print(combined_df.describe())
```
